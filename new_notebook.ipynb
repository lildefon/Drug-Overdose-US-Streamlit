```python
provisional_death['Year'] = provisional_death['Year'].astype(str)
provisional_death["Month"]= provisional_death['Month'].astype(str)



def parse_date(date_string):
    try:
        # Try parsing with format='%Y-%b' for year-month abbreviation (e.g., 2015-Apr)
        return pd.to_datetime(date_string, format='%Y-%b')
    except ValueError:
        try:
            # If the first attempt fails, try parsing with format='%Y-%B' for full month name (e.g., 2015-April)
            return pd.to_datetime(date_string, format='%Y-%B')
        except ValueError:
            # If both attempts fail, return NaT (Not a Time)
            return pd.NaT

provisional_death["Date"] = provisional_death["Year"] + "-" + provisional_death["Month"]

# Apply the custom parsing function to each element in the 'Date' column
provisional_death["Date"] = provisional_death["Date"].apply(parse_date)





```
